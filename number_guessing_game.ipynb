## Level:1 Giving multiple conditions
1. user shld only give at the range(1,100)
2. the prog shld say lesser or lower for the program

```python
print("Welcome to the nunber guessing game")
secret_num = 34
for i in range(0,7):
    guess_number = int(input("Enter ur guessed number"))
    if guess_number < 0 or guess_number > 100:
        print("Enter the number greater than 0 and less than 100")
    elif guess_number == secret_num:
        print("U guessed the correct number")
        print("the number tries u took is",i+1)
        break
    elif guess_number < secret_num:
        print("The number is greater than",guess_number)
    elif guess_number > secret_num:
        print("The number is lesser than the ",guess_number)
```
