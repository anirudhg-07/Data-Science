### 1. List
- It can take duplicate elements
- Heterogenious: It can take the all types of elements
- 1 Dimension
- Indexing and slicing is possible
- Mutable - The value can be changed in the future

```python
empty_cart1 = []
```

```python
items_to_buy = ["Apple", "Orange","Grapes","Apple",9,7.66]
items_to_buy
```

```python
items_to_buy[0:3:2]
```

```python
items_to_buy[2] = "Watermelon"
```

```python
items_to_buy
```

### 2. Sets
- 1D
- Heterogenious
- It does not have any duplicate values
- Slicing and Indexing is not possible
- Mutable

```python
empty_cart2 = set()
```

```python
type(empty_cart2)
```

```python
items_to_buy2 ={"Apple", "Orange","Grapes","Apple",9,7.66}
items_to_buy2
```

```python
items_to_buy2.add("Banana")
```

```python
items_to_buy2
```
