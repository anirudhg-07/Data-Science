### 4 Data structure
* List
* Tuple
* Sets
* Dictionary - Key & value


### 1. List
- It can take duplicate elements
- Heterogenious: It can take the all types of elements
- 1 Dimension
- Indexing and slicing is possible
- Mutable - The value can be changed in the future

```python
empty_cart1 = []
```

```python
items_to_buy = ["Apple", "Orange","Grapes","Apple",9,7.66]
items_to_buy
```

```python
items_to_buy[0:3:2]
```

```python
items_to_buy[2] = "Watermelon"
```

```python
items_to_buy
```

### 2. Sets
- 1D
- Heterogenious
- It does not have any duplicate values
- Slicing and Indexing is not possible
- Mutable

```python
empty_cart2 = set()
```

```python
type(empty_cart2)
```

```python
items_to_buy2 ={"Apple", "Orange","Grapes","Apple",9,7.66}
items_to_buy2
```

```python
items_to_buy2.add("Banana")
```

```python
items_to_buy2
```

### 3. Tuple
- 1D
- Heterogenious
- Allows Duplicate values
- Slicing and indexing is possible
- Immutable - Once the value is appended the values cant be changed

```python
empty_cart3 = ()
```

```python
type(empty_cart3)
```

```python
items_to_buy3 = ("Apple", "Orange","Grapes","Apple",9,7.66)
```

```python
items_to_buy3[0]
```

```python
items_to_buy3.count("Apple")
```

### 4. Dictionary

- It is a key value pair
- 2d
- Heterogenoius
- Mutable
- Indexing and slicing is possible
- Will not create duplicate key

```python
empty_cart4 = {}
```

```python
type(empty_cart4)
```

```python
student_marks = {"Maths":98,"English":87,"Tamil":99,}
student_marks
```

```python
student_marks.keys()
```

```python
student_marks.values()
```
